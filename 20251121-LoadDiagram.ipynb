```python
import numpy as np

L = 1.0
F = 1000.0
a = 0.7 * L

x = np.linspace(0, L, 300)
mask = x <= a

V = np.zeros_like(x)
V[mask] = -F

M = np.zeros_like(x)
M[mask] = -F * (a - x[mask])

V_at_fixed = V[0]
M_max = M[0]
V_drop_point = V[np.argmax(x >= a)]

print("=== Beam Results ===")
print(f"Beam length L           = {L} m")
print(f"Load F                  = {F} N")
print(f"Load position a         = {a} m")
print()
print("=== Key Values ===")
print(f"Shear at fixed end      = {V_at_fixed:.2f} N")
print(f"Bending moment at wall  = {M_max:.2f} N·m")
print(f"Shear just after load   = {V_drop_point:.2f} N")
print()
print("Arrays ready: x, V, M")
```
